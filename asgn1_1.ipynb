```python
import numpy as np
from matplotlib import pyplot as plt
import random
%matplotlib inline
from IPython import display
import time


# Sigmoid Function
def sigmoid(u):
    return 1 / (1 + np.exp(-u))

#intilaizing values for NAND gate

x1= np.array([0,0,1,1]) #input1
x2= np.array([0,1,0,1]) #input2
t = np.array([1,1,1,0]) #output
error_list = []

#given random values range from -2,2

w1= random.uniform(-2,2)
print(w1)
w2=random.uniform(-2,2)
print(w2)
w3= 1 #bias
#learning rate
le = 0.5
k = np.linspace(-2 ,2 ,3)

print("   ") 

for i in range(0,4):
  if t[i]>0:
    plt.scatter(x1[i],x2[i],marker ="o" ,c="b")
  else:
    plt.scatter(x1[i],x2[i],marker ="s" ,c="r")
while True:
    total_error=0
    for i in range(4):
        #neuron equation
        u = (x1[i]*w1)+(x2[i]*w2)+w3
        
        #activation function:sigmoid
        y = sigmoid(u) 
        print(u)
        print(y)
        print("   ")
        
        # error function
        E = np.square(t[i]- y)
        print("E" ,E ,i)
        
        #derivative of w1
        dw1 = le * (t[i]- y) * y * (1-y) * x1[i]
        #derivative of w2
        dw2 = le * (t[i]- y) * y * (1-y) * x2[i]
        #derivative of w3
        dw3 = le * (t[i]- y) * y * (1-y)

        #for loop ends

    total_error = E + total_error
    print("total error",total_error)
    # change the weights
        
    w1 = w1 + dw1
    w2 = w2 + dw2 
    w3 = w3 + dw3
    #drawing line
    l = -((w3 + (k*w1))/w2) # line equation to plot

    plt.plot(k,l,linestyle='--', c='g') 

    display.display(plt.gcf()) 

    display.clear_output(wait=True)

    time.sleep(0.000001)



    error_list.append(total_error)
    print("elist",error_list)
    
    if total_error < 0.01:
      break; 
   
      
    
        
        
        
        
        
        
        
           
        
    






```

```python
plt.plot( range(len(error_list)) ,error_list)
plt.title("Line Graph")
plt.ylabel("Error List")
plt.xlabel("Epochs")
plt.show  
```
